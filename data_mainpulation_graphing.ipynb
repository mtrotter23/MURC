```python
import pandas as pd 
import altair as alt 
```

```python
#reading in data
bear = pd.read_csv('bb.csv')

#weather on Vancouver Island as a Whole
van_island=pd.read_csv('van_islandclimate-daily.csv', low_memory=False)
#weather in BC ass a whole 
bc = bc = pd.read_csv('BC_climate_daily.csv', low_memory=False)


#Plan
#filter each data set out year, 
#group they by 'LOCAL_MONTH', with mean
#pd.concat them all together 
#compare all the weather patterns to see the relationship between them all. -> aprove of trend hopefully!! 
```

```python
precip_vi = van_island[['LOCAL_YEAR', 'LOCAL_DAY', 'LOCAL_MONTH', 'TOTAL_PRECIPITATION', 'TOTAL_RAIN', 'TOTAL_SNOW']].dropna()
precip_bc = bc[['LOCAL_YEAR', 'LOCAL_DAY', 'LOCAL_MONTH', 'TOTAL_PRECIPITATION', 'TOTAL_RAIN', 'TOTAL_SNOW']].dropna()

precip_vi_2011 = precip_vi[precip_vi['LOCAL_YEAR']==2011].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'VI')
precip_vi_2012 = precip_vi[precip_vi['LOCAL_YEAR']==2012].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'VI')
precip_vi_2013 = precip_vi[precip_vi['LOCAL_YEAR']==2013].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'VI')
precip_vi_2014 = precip_vi[precip_vi['LOCAL_YEAR']==2014].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'VI')
precip_vi_2015 = precip_vi[precip_vi['LOCAL_YEAR']==2015].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'VI')
precip_vi_2016 = precip_vi[precip_vi['LOCAL_YEAR']==2016].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'VI')
precip_vi_2017 = precip_vi[precip_vi['LOCAL_YEAR']==2017].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'VI')
precip_vi_2018 = precip_vi[precip_vi['LOCAL_YEAR']==2018].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'VI')
precip_vi_2019 = precip_vi[precip_vi['LOCAL_YEAR']==2019].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'VI')
precip_vi_2020 = precip_vi[precip_vi['LOCAL_YEAR']==2020].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'VI')
precip_vi_2021 = precip_vi[precip_vi['LOCAL_YEAR']==2021].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'VI')
precip_vi_2022 = precip_vi[precip_vi['LOCAL_YEAR']==2022].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'VI')
precip_vi_2023 = precip_vi[precip_vi['LOCAL_YEAR']==2023].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'VI')
#precip_vi_2024 = precip_vi[precip_vi['LOCAL_YEAR']==2024].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'VI')

precip_vi_month=pd.concat([precip_vi_2011,
                       precip_vi_2012, 
                       precip_vi_2013,
                       precip_vi_2014,
                       precip_vi_2015,
                       precip_vi_2016,
                       precip_vi_2017,
                       precip_vi_2018,
                       precip_vi_2019,
                       precip_vi_2020,
                       precip_vi_2021,
                       precip_vi_2022,
                       precip_vi_2023
                       ], ignore_index=True, axis=0)

precip_bc_2011 = precip_bc[precip_bc['LOCAL_YEAR']==2011].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'BC')
precip_bc_2012 = precip_bc[precip_bc['LOCAL_YEAR']==2012].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'BC')
precip_bc_2013 = precip_bc[precip_bc['LOCAL_YEAR']==2013].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'BC')
precip_bc_2014 = precip_bc[precip_bc['LOCAL_YEAR']==2014].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'BC')
precip_bc_2015 = precip_bc[precip_bc['LOCAL_YEAR']==2015].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'BC')
precip_bc_2016 = precip_bc[precip_bc['LOCAL_YEAR']==2016].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'BC')
precip_bc_2017 = precip_bc[precip_bc['LOCAL_YEAR']==2017].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'BC')
precip_bc_2018 = precip_bc[precip_bc['LOCAL_YEAR']==2018].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'BC')
precip_bc_2019 = precip_bc[precip_bc['LOCAL_YEAR']==2019].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'BC')
precip_bc_2020 = precip_bc[precip_bc['LOCAL_YEAR']==2020].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'BC')
precip_bc_2021 = precip_bc[precip_bc['LOCAL_YEAR']==2021].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'BC')
precip_bc_2022 = precip_bc[precip_bc['LOCAL_YEAR']==2022].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'BC')
precip_bc_2023 = precip_bc[precip_bc['LOCAL_YEAR']==2023].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'BC')
#precip_bc_2024 = precip_bc[precip_bc['LOCAL_YEAR']==2024].groupby('LOCAL_YEAR').mean(numeric_only=True).reset_index().assign(region = 'BC')

precip_bc_month= pd.concat([precip_bc_2011,
                            precip_bc_2012,
                            precip_bc_2013,
                            precip_bc_2014,
                            precip_bc_2015,
                            precip_bc_2016,
                            precip_bc_2017,
                            precip_bc_2018,
                            precip_bc_2019,
                            precip_bc_2020,
                            precip_bc_2021,
                            precip_bc_2022,
                            precip_bc_2023], ignore_index=True, axis=0)

precip_both = pd.concat([precip_bc_month, precip_vi_month], ignore_index=True, axis=0)
compare_precip_both =alt.Chart(precip_both).mark_line().encode(
    x=alt.X('LOCAL_YEAR').scale(zero=False),
    y=alt.Y('TOTAL_PRECIPITATION').scale(zero=False),
    color='region'
)
#shows simillar trend in weather patterns between vancouver island and bc, thus can make similar trends in how Vancouver island 
compare_precip_both
bb_year=bear.groupby('year').mean(numeric_only=True).reset_index()
bear_plot = alt.Chart(bb_year).mark_line().encode(
    x='year', 
    y='calls_received'
)
display(bear_plot, compare_precip_both)
```
